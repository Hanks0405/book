```python
%%tikz -p kotex,pgfplots -l arrows.meta,calc,positioning,shapes -g fillbetween -s 2000,1200 -f png

\pgfplotsset{compat=1.15}
\coordinate (leftbottom) at (0,0);
\coordinate (righttop) at (20,17);
\fill[white,use as bounding box] (leftbottom) rectangle (righttop);
\draw[help lines,white] (leftbottom) grid (righttop);
\tikzset{
    >=stealth,font={\ttfamily\normalsize}
}

\begin{scope}[shift={(0,8.5)}]
    \pgfplotsset{width=20cm, height=9cm}
    \begin{axis}[hide axis, ymin=0, ymax=1]
    \addplot[line width=1pt, domain=-9:9, samples=100] {
        1 / (1 + exp(-x))
    };
    \addplot[line width=1pt, domain=-0.5:2, samples=100, dashed] {
        1 / (1 + exp(-x))
    } \closedcycle ;
    \addplot[name path=line, line width=1.7pt, no markers, domain=-9:9] {0};
    \pgfmathsetmacro{\na}{1.0 / (1.0 + exp(0.5))}
    \pgfmathsetmacro{\nb}{1.0 / (1.0 + exp(-2))}
    \fill (-0.5, \na) circle (2pt) coordinate (p11);
    \fill (2, \nb) circle (2pt) coordinate (p21);
    \node at (-7, 0.1) {누적분포함수};
    \end{axis}
\end{scope}

\begin{scope}[shift={(0,2)}]
    \pgfplotsset{width=20cm, height=9cm}
    \begin{axis}[hide axis, ymin=0, ymax=1]
    \addplot[name path=sigmoid, line width=1pt, domain=-9:9, samples=100] {
        3 / ((1 + exp(-x)) * (1 + exp(x)))
    };
    \addplot[name path=sigmoid, line width=1pt, domain=-0.5:2, samples=100, fill=gray] {
        3 / ((1 + exp(-x)) * (1 + exp(x)))
    } \closedcycle ;
    \addplot[name path=line, line width=1.7pt, no markers, domain=-9:9] {0};

    \fill (-0.5, 0) circle (2pt) coordinate (p12);
    \fill (2, 0) circle (2pt) coordinate (p22);
    \node at (-7, 0.1) {확률밀도함수};
    \end{axis}
\end{scope}

\draw[dashed] (p11) -- (p12);
\draw[dashed] (p21) -- (p22);
\node[anchor=south east] at (p11) {$F(x_1)$};
\node[anchor=south east] at (p21) {$F(x_2)$};
\node[below] at (p12) {$x_1$};
\node[below] at (p22) {$x_2$};
\fill (p11 -| p21) coordinate (p31) circle (2pt);
\draw[dashed] (p11) -- (p31);
\draw[dotted] (p21) -- +(1,0);
\draw[dotted] (p31) -- +(1,0);
\draw[<->] ($(p21)+(0.5,0)$) -- ($(p31)+(0.5,0)$);
\node[right=0.3cm, fill=white] at ($(p21)!0.5!(p31)$) {$F(x_2) - F(x_1)$};
```
