```python
A1 = set([1, 2, 3, 4])
A2 = set([2, 4, 6])
A3 = set([1, 2, 3])
A4 = set([2, 3, 4, 5, 6])
```

```python
A1.union(A2)
```

```python
A2 | A1
```

```python
A3.intersection(A4)
```

```python
A4 & A3
```

```python
A3.issubset(A1)
```

```python
A3 <= A1
```

```python
A3.issubset(A2)
```

```python
A3 <= A2
```

```python
A3 <= A3  # 모든 집합은 자기 자신의 부분집합이다.
```

```python
A3 < A3  # 모든 집합은 자기 자신의 진부분집합이 아니다.
```

```python
A1.difference(A2)
```

```python
A1 - A2
```

```python
empty_set = set([])
empty_set
```

```python
empty_set < A1
```

```python
empty_set.intersection(A1)
```
