#### 연습 문제 1.2.2의 답

(1)

$$
\begin{align}
\begin{aligned}
\prod_i x_i^{y_i} \\
= & \; x_1^{y_1} \cdot  x_2^{y_2} \cdot x_3^{y_3} \cdot x_4^{y_4} \\
= & \; x_1^0 \cdot  x_2^1 \cdot x_3^0 \cdot x_4^0 \\
= & \; 1 \cdot  x_2 \cdot 1 \cdot 1 \\
= & \; x_2 
\end{aligned}
\tag{A.1.5}
\end{align}
$$

(2)

$$
\begin{align}
\begin{aligned}
\prod_i x_i^{y_i} \\
= & \; x_1^{y_1} \cdot  x_2^{y_2} \cdot x_3^{y_3} \cdot x_4^{y_4} \\
= & \; x_1^0 \cdot  x_2^0 \cdot x_3^1 \cdot x_4^0 \\
= & \; 1 \cdot 1 \cdot x_3 \cdot 1 \\
= & \; x_3
\end{aligned}
\tag{A.1.6}
\end{align}
$$


#### 연습 문제 1.2.3의 답


(1)

$$
\begin{align}
\begin{aligned}
\left( \sum_{i=1}^3 x_i \right)^2 \\
=  \big( x_1 + x_2 + & x_3 \big)^2  \\
=  \big( x_1 + x_2 + & x_3 \big) \cdot \big( x_1 + x_2 + x_3 \big) \\
=  x_1 x_1 +  x_1 x_2 + & x_1 x_3 + x_2 x_1 + x_2 x_2 + x_2 x_3 + x_3 x_1 + x_3 x_2 + x_3 x_3 \\
\end{aligned}
\tag{A.1.7}
\end{align}
$$

$$
\begin{align}
\begin{aligned}
\sum_{i=1}^3 \sum_{j=1}^3 x_i x_j & \\
= x_1 x_1 +  x_1 x_2 + & x_1 x_3 + x_2 x_1 + x_2 x_2 + x_2 x_3 + x_3 x_1 + x_3 x_2 + x_3 x_3 
\end{aligned}
\tag{A.1.8}
\end{align}
$$

$$
\begin{align}
\begin{aligned}
\therefore \;\;\;
\left( \sum_{i=1}^3 x_i \right)^2 = \sum_{i=1}^3 \sum_{j=1}^3 x_i x_j
\end{aligned}
\tag{A.1.9}
\end{align}
$$

(2)

$$
\begin{align}
\begin{aligned}
\sum_{i=1}^3 \sum_{j=1}^3 x_i y_{ij} & \\ 
= x_1 y_{11} + x_1 y_{12} & + x_1 y_{13} + x_2 y_{21} + x_2 y_{22} + x_2 y_{23} + x_3 y_{31} + x_1 y_{32} + x_3 y_{33}
\end{aligned}
\tag{A.1.10}
\end{align}
$$

$$
\begin{align}
\begin{aligned}
\sum_{i=1}^3 \left( x_i  \sum_{j=1}^3 y_{ij} \right) & \\
= \sum_{i=1}^3 \big( x_i  ( y_{i1} + y_{i2} + & y_{i3} ) \big) \\
= \sum_{i=1}^3 \big( x_i y_{i1} + x_i y_{i2} + & x_i y_{i3} \big) \\
= x_1 y_{11} + x_1 y_{12}  + x_1 y_{13} + & x_2 y_{21} + x_2 y_{22} + x_2 y_{23} + x_3 y_{31} + x_1 y_{32} + x_3 y_{33}
\end{aligned}
\tag{A.1.11}
\end{align}
$$

$$
\begin{align}
\begin{aligned}
\therefore \;\;\;
\sum_{i=1}^3 \sum_{j=1}^3 x_i y_{ij} =  \sum_{i=1}^3 \left( x_i  \sum_{j=1}^3 y_{ij} \right)
\end{aligned}
\tag{A.1.12}
\end{align}
$$


#### 연습 문제 1.2.4의 답

$$
\begin{align}
\begin{aligned}
\prod_{i=1}^2 \sum_{X_i} x_{i} \\
=& \sum_{X_1} x_{1}  \cdot \sum_{X_2} x_{2} \\
=& (x_{1,1} + x_{1,2} + x_{1,3}) \cdot (x_{2,1} + x_{2,2} + x_{2,3}) \\
=& x_{1,1}(x_{2,1} + x_{2,2} + x_{2,3}) \\
+& x_{1,2}(x_{2,1} + x_{2,2} + x_{2,3}) \\
+& x_{1,3}(x_{2,1} + x_{2,2} + x_{2,3}) \\
=& x_{1,1}x_{2,1} + x_{1,1}x_{2,2} + x_{1,1}x_{2,3} + \\
 & x_{1,2}x_{2,1} + x_{1,2}x_{2,2} + x_{1,2}x_{2,3} + \\ 
 & x_{1,3}x_{2,1} + x_{1,3}x_{2,2} + x_{1,3}x_{2,3}
\end{aligned}
\tag{A.1.13}
\end{align}
$$

$$
\begin{align}
\begin{aligned}
\sum_{X_1 \times X_2} \prod_{i=1}^2 x_i \\
=& \sum_{X_1 \times X_2} x_1 x_2 \\
=& x_{1,1}x_{2,1} + x_{1,1}x_{2,2} + x_{1,1}x_{2,3} + \\
 & x_{1,2}x_{2,1} + x_{1,2}x_{2,2} + x_{1,2}x_{2,3} + \\ 
 & x_{1,3}x_{2,1} + x_{1,3}x_{2,2} + x_{1,3}x_{2,3}
\end{aligned}
\tag{A.1.14}
\end{align}
$$

$$
\begin{align}
\therefore \;\;\;
\prod_{i=1}^2 \sum_{X_i} x_{i} =
\sum_{X_1 \times X_2} \prod_{i=1}^2 x_i
\tag{A.1.15}
\end{align}
$$
