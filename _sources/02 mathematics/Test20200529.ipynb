## 10.

동전이 있다. 이 동전을 던져 앞면이 나올 확률은 베르누이 확률분포의 모수 $\mu$로 표현할 수 있고 $\mu$값의 추정 결과는 베이지안 추정법을 사용하여 베타 분포의 모수로 표현할 수 있다. 동전을 던지기 전에는 동전에 대해 아는 바가 전혀 없다고 가정하면 동전을 10번 던져서 6번 앞면이 나온 경우 베타 분포의 모수는 얼마인가?



### 답:

동전을 던져 앞면이 나온 횟수를 $N_1$, 뒷면이 나온 횟수를 $N_0$라고 하면 관측 이전의 베타 분포 모수 $a, b$와 관측 이후의 베타 분포 모수 $a', b'$는 다음과 같은 관계가 있다.

$$
\begin{eqnarray}
a' &=& N_1 + a \\
b' &=& N_0 + b 
\end{eqnarray}
$$

동전을 던지기 전에는 동전에 대해 아는 바가 전혀 없으므로 $a=b=1$이다. 따라서

$$
\begin{eqnarray}
a' &=& 6 + 1 = 7 \\
b' &=& 4 + 1 = 5
\end{eqnarray}
$$

이다.

1이 나올 확률 파라미터 $\theta$를 가지는 베르누이 분포에서 $N$개의 샘플 $x_1, x_2, \cdots, x_N$을 뽑았을 때, 이 값에서 최대가능도 방법으로 $\theta$를 추정하고자 합니다. 로그가능도(Log-Likelihood)를 구하세요


* 참조: Bernoulli 분포

$$ P(x | \theta ) = \text{Bern}(x | \theta ) = \theta^x (1 - \theta)^{1-x}$$


### 답:



* 가능도

$$ L = P(x_{1:N}|\theta) = \prod_{i=1}^N  \theta^{x_i} (1 - \theta)^{1-x_i} $$

* 로그가능도

$$ 
\begin{eqnarray*}
\log L 
&=& \log P(x_{1:N}|\theta)  \\
&=& \sum_{i=1}^N \big\{ {x_i} \log\theta + (1-x_i)\log(1 - \theta) \big\} \\
&=&  \sum_{i=1}^N {x_i} \log\theta +  \left( N-\sum_{i=1}^N x_i \right) \log( 1 - \theta ) \\
\end{eqnarray*}
$$
