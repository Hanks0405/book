#### 연습 문제 9.2.1의 답

로그가능도함수는 다음과 같다.

$$
\begin{align}
\begin{aligned}
& \log L(\mu; x_1=1, x_2=0, x_3=1, x_4=1) \\
&= \log  p(x_1=1, x_2=0, x_3=1, x_4=1; \mu) \\
&= \log \big( p(x=1;\mu)p(x=0;\mu)p(x=1;\mu)p(x=1;\mu) \big) \\
&= \log \big( \mu \cdot (1-\mu) \cdot \mu  \cdot \mu \big) \\
&= \log \big( \mu^3(1-\mu)\big)
\end{aligned}
\tag{A.9.1}
\end{align}
$$

이 로그가능도함수를 최대화하는 모수의 값을 찾기위해 미분한 도함수가 0이 되는 위치를 찾는다. 

$$
\begin{align}
\dfrac{d\log L}{d\mu} = \frac{3\mu^2 - 4\mu^3}{\mu^3(1-\mu)} = 0
\tag{A.9.2}
\end{align}
$$

$$
\begin{align}
\mu = \frac{3}{4}
\tag{A.9.3}
\end{align}
$$


#### 연습 문제 9.2.2의 답

$K=4$인 카테고리분포의 확률질량함수는 다음과 같다.


$$
\begin{align}
P(x_; \mu) = \mu_1^{x_1} \mu_2^{x_2} \mu_3^{x_3} \mu_4^{x_4}
\tag{A.9.4}
\end{align}
$$

따라서 전체 데이터에 대한 로그가능도함수는

$$
\begin{align}
\begin{aligned}
& \log L(\mu; x_1=1, x_2=4, x_3=1, x_4=2, x_5=4, x_6=2, x_7=3, x_8=4) \\
&= \log  p(x_1=1, x_2=4, x_3=1, x_4=2, x_5=4, x_6=2, x_7=3, x_8=4; \mu) \\
&= \log \big( p(x=1;\mu)p(x=4;\mu)p(x=1;\mu)p(x=2;\mu)p(x=4;\mu)p(x=2;\mu)p(x=3;\mu)p(x=4;\mu) \big) \\
&= \log \mu_1 \mu_4 \mu_1 \mu_2 \mu_4 \mu_2 \mu_3 \mu_4 \\
&= \log \mu_1^2 \mu_2^2 \mu_3 \mu_4^3 \\
&= 2\log \mu_1 + 2\log \mu_2 + \log \mu_3 + 3\log \mu_4
\end{aligned}
\tag{A.9.5}
\end{align}
$$

그런데 이 모수값은 다음 제한조건을 만족해야 한다.

$$
\begin{align}
\mu_1 + \mu_2 + \mu_3 + \mu_4 - 1 = 0
\tag{A.9.6}
\end{align}
$$

이 제한조건을 만족하면서 로그가능도함수를 최대화하는 모수의 값을 찾기위해 라그랑주방법을 사용하여 새로운 목적함수 $J$를 만든다.

$$
\begin{align}
J = 2\log \mu_1 + 2\log \mu_2 + \log \mu_3 + 3\log \mu_4 + \lambda(\mu_1 + \mu_2 + \mu_3 + \mu_4 - 1)
\tag{A.9.7}
\end{align}
$$

목적함수 $J$를 모수와 라그랑주승수로 미분하여 그레디언트가 0이 되는 값을 찾는다.

$$
\begin{align}
\begin{aligned}
\dfrac{\partial J}{\partial \mu_1} &= \dfrac{2}{\mu_1} + \lambda = 0 \\
\dfrac{\partial J}{\partial \mu_2} &= \dfrac{2}{\mu_2} + \lambda = 0 \\
\dfrac{\partial J}{\partial \mu_3} &= \dfrac{1}{\mu_3} + \lambda = 0 \\
\dfrac{\partial J}{\partial \mu_4} &= \dfrac{3}{\mu_4} + \lambda = 0 \\
\dfrac{\partial J}{\partial \lambda} &= \mu_1 + \mu_2 + \mu_3 + \mu_4 - 1 = 0 \\
\end{aligned}
\tag{A.9.8}
\end{align}
$$

이 연립방정식을 풀면

$$
\begin{align}
\mu_1 = \frac14, \;\;\mu_2 = \frac14, \;\;\mu_3 = \frac18, \;\;\mu_4 = \frac38
\tag{A.9.9}
\end{align}
$$

#### 연습 문제 9.3.1의 답

사전분포가 하이퍼모수  $a=b=1$인 베타분포인 경우 사후분포는 다음과 같은 하이퍼모수를 가지는 베타분포가 된다.

상품 A의 사후분포의 하이퍼모수:

$$
\begin{align}
\begin{aligned}
a'_A &= 2 + 1 = 3 \\
b'_A &= 1 + 1 = 2
\end{aligned}
\tag{A.9.10}
\end{align}
$$

상품 B의 사후분포의 하이퍼모수:

$$
\begin{align}
\begin{aligned}
a'_B &= 60 + 1 = 61 \\
b'_B &= 40 + 1 = 41
\end{aligned}
\tag{A.9.11}
\end{align}
$$

```python
aA, bA = 3, 2
aB, bB = 61, 41
xx = np.linspace(0, 1, 1000)
plt.plot(xx, sp.stats.beta(aA, bA).pdf(xx), 
         c="g", ls="-", label="상품 A의 모수")
plt.plot(xx, sp.stats.beta(aB, bB).pdf(xx), 
         c="r", ls="--", label="상품 B의 모수")
plt.legend()
plt.title("베이즈 추정법으로 계산한 베르누이분포 모수의 사후분포")
plt.show()
```

#### 연습 문제 9.3.2의 답

사이킷런에서 제공하는 붓꽃데이터 중 꽃받침의 길이를 정규분포라고 가정하고 베이즈 추정법을 사용하여 기댓값을 추정하라. 단, 정규분포의 분산으로는 표본분산을 사용하고 

사전분포가 기댓값 1, 분산 1인 정규분포이므로

$$
\begin{align}
\begin{aligned}
\mu_0 &= 1 \\
\sigma_0^2 &= 1
\end{aligned}
\tag{A.9.12}
\end{align}
$$

사이킷런에서 제공하는 붓꽃데이터 중 꽃받침의 길이의 표본평균과 표본분산을 구하면 5.84와 0.68이다.

```python
from sklearn.datasets import  load_iris
iris = load_iris()
x = iris.data[:, 0]
N = len(x)
mu = np.mean(x)
sigma2 = np.var(x)
print("sample mean:", mu)
print("sample variance:", sigma2)
sns.distplot(x, kde=False, fit=sp.stats.norm)
plt.vlines(x=mu, ymin=0, ymax=0.6, ls="--")
plt.vlines(x=mu+2*np.sqrt(sigma2), 
           ymin=0, ymax=0.6, ls=":")
plt.vlines(x=mu-2*np.sqrt(sigma2), 
           ymin=0, ymax=0.6, ls=":")
plt.show()
```

베이즈 추정법을 사용하여 사후분포를 구하면 다음과 같다.

$$
\begin{align}
\begin{aligned}
\mu'_0 &= \dfrac{\sigma^2}{N\sigma_0^2 + \sigma^2}\mu_0 + \dfrac{N\sigma_0^2}{N\sigma_0^2 + \sigma^2} \dfrac{\sum x_i}{N} \approx 5.82 \\
\sigma_0^{'2} &= \left( \dfrac{1}{\sigma_0^{2}} + \dfrac{N}{\sigma^{2}} \right)^{-1} \approx 0.0045
\end{aligned}
\tag{A.9.13}
\end{align}
$$

```python
mu0 = 1
sigma02 = 1

mu_post = sigma2 / (N * sigma02 + sigma2) * mu0 \
          + (N * sigma02) / (N * sigma02 + sigma2) * mu
sigma2_post = 1 / (1 / sigma02 + N / sigma2)
print("post meam:", mu_post)
print("post variance:", sigma2_post)
```

```python
plt.subplot(211)
sns.distplot(x, kde=False, fit=sp.stats.norm)
plt.vlines(x=mu_post, ymin=0, ymax=0.6, ls="--")
plt.xlim(4, 8)
plt.subplot(212)
xx = np.linspace(4, 8, 1000)
plt.plot(xx, 
         sp.stats.norm(loc=mu_post, 
                       scale=np.sqrt(sigma2_post)).pdf(xx))
plt.title("기댓값의 사후분포")
plt.xlim(4, 8)
plt.tight_layout()
plt.show()
```

어떤 인터넷 쇼핑몰의 상품 20개의 상품평이 있고 '좋아요'가 11개 또는 '싫어요'가 9개다. 이 상품이 좋다는 주장을 검정하라. 유의수준은 10%다.

검정통계량은 좋아요 갯수 11이고 이 검정통계량은 $N=11+9=20$인 이항분포를 따른다.

$$
\begin{align}
\text{유의확률} = 1 - F(t_0 - 1) = 1 - F(10) = 0.412
\tag{A.9.16}
\end{align}
$$

```python
N = 20
mu = 0.5
rv = sp.stats.binom(N, mu)
1 - rv.cdf(11 - 1)
```

유의확률이 41.2%이므로 유의수준 10%보다 높고 귀무가설을 기각할 수 없다. 따라서 상품평이 좋다고 확정할 수 없다. 

```python
xx = np.arange(N + 1)

plt.subplot(211)
plt.stem(xx, rv.pmf(xx))
plt.ylabel("pmf")
plt.title("검정통계량분포(N=20인 이항분포)의 확률질량함수")
black = {"facecolor": "black"}
plt.annotate('검정통계량 t=11', xy=(11, 0.0), 
             xytext=(11, 0.05), arrowprops=black)

plt.subplot(212)
plt.stem(xx, rv.cdf(xx))
plt.ylabel("cdf")
plt.title("검정통계량분포(N=20인 이항분포)의 누적분포함수")
plt.annotate('검정통계량 t=11', xy=(11, 0.0), 
             xytext=(11, 0.3), arrowprops=black)

plt.tight_layout()
plt.show()
```

```python
N1 = 10
pvalues1 = np.array([sp.stats.binom_test(n, N1) \
                     for n in range(N1 + 1)])
fair_range1, = np.where(pvalues1 > 0.1)
fair_start1, fair_end1 = fair_range1[0], fair_range1[-1]
print("N={}일 때 공정한 경우: 앞면이 나온 횟수가 {}~{}일 때"\
      .format(N1, fair_start1, fair_end1))

N2 = 1000
pvalues2 = np.array([sp.stats.binom_test(n, N2) \
                     for n in range(N2 + 1)])
fair_range2, = np.where(pvalues2 > 0.1)
fair_start2, fair_end2 = fair_range2[0], fair_range2[-1]
print("N={}일 때 공정한 경우: 앞면이 나온 횟수가 {}~{}일 때"\
      .format(N2, fair_start2, fair_end2))
```

```python
plt.subplot(211)
plt.plot(pvalues1, "r-")
plt.hlines(y=0.1, xmin=0, 
           xmax=N1 + 1, ls="--")
plt.xlim(0, N1)
plt.title("N={}일 때 유의확률".format(N1))
plt.subplot(212)
plt.plot(pvalues2, "r-")
plt.hlines(y=0.1, xmin=0, 
           xmax=N2 + 1, ls="--")
plt.xlim(0, N2)
plt.title("N={}일 때 유의확률".format(N2))
plt.xlabel("앞면이 나온 횟수")
plt.tight_layout()
plt.show()
```

#### 연습 문제 9.5.3의 답

이 문제는 모수가 $\mu=\frac23$인 이항검정을 사용하여 풀 수 있다. 단 우측검정을 사용해야 한다. 다음 그래프에서 응답자 수가 1116명 이상이면 70%만 찬성해도 모수가 $\mu=\frac23$라는 결론을 얻을 수 있다.

```python
pvalues = []
N_range = range(1050, 1180)
for N in N_range:
    n = int(round(0.7 * N))
    pvalues.append(
        sp.stats.binom_test(n, N, p=2/3, 
                            alternative="greater"))

plt.plot(N_range, pvalues, "ro-")
plt.hlines(y=0.01, xmin=1050, xmax=1180, ls="--")
plt.xlim(1050, 1180)
plt.xlabel("응답자 수")
plt.ylabel("유의확률")
plt.title("응답자 수에 대한 유의확률의 변화")
plt.show()
```

#### 연습 문제 9.5.5의 답

두 카테고리 분포의 모수를 비교검정하려면 `chi2_contingency()`를 사용한다. 모수가 같은 것이 디폴트 귀무가설이므로 다음과 같이 유의확률을 계산할 수 있다. 유의확률이 0.7%이므로 두 분포의 모수가 다르다. 즉, 데이터 사이언스 스쿨 수업을 들었는가의 여부가 머신러닝 수업의 학점과 상관관계가 있다고 말할 수 있다.

```python
obs = np.array([[4, 16, 20], [23, 18, 19]])
result = sp.stats.chi2_contingency(obs)
print(result[1])
```

#### 연습 문제 9.5.6의 답

두 분포의 평균값이 같은지 비교하려면 `ttest_ind()` 함수를 사용한다. 유의확률이 54%이므로 1반의 실력이 2반보다 좋다고 이야기 할 수 없다.

```python
x1 = [80, 75, 85, 50, 60, 75, 45, 70, 90, 95, 85, 80]
x2 = [80, 85, 70, 80, 35, 55, 80]
sp.stats.ttest_ind(x1, x2, equal_var=True)
```

두 표본집합의 분산이 다르다고 가정할 경우도 결론은 마찬가지다.

```python
sp.stats.ttest_ind(x1, x2, equal_var=False)
```
