#### 연습 문제 3.2.2의 답 

만약 벡터 $v_1, \ldots, v_N$에 대해 선형종속을 만드는 모두 0이 아닌 계수 $c_1, \ldots, c_N$이 존재한다면

$$
\begin{align}
c_1v_1 + \cdots + c_Nv_N = 0
\tag{A.3.38}
\end{align}
$$

이 성립하고 여기에 $v_i \, (i = 1, \ldots, N)$을 내적하면

$$
\begin{align}
v_i(c_1v_1 + \cdots + c_Nv_N) = 0
\tag{A.3.39}
\end{align}
$$

가 성립한다.

그런데 이 벡터들은 서로 직교하므로 $j\neq i$인 모든 $j$에 대해 

$$
\begin{align}
v_i^Tv_j = 0
\tag{A.3.40}
\end{align}
$$

이다.

따라서 

$$
\begin{align}
v_i(c_1v_1 + \cdots + c_Nv_N) = c_iv_i^Tv_i = c_i \|v_i\|^2 = 0
\tag{A.3.41}
\end{align}
$$

이 성립한다.

그런데 이 벡터들은 영벡터가 아니므로 

$$
\begin{align}
c_i = 0
\tag{A.3.42}
\end{align}
$$

이고 이는 모든 $i=1,\ldots,N$에 대해 성립하므로 선형종속을 만드는 모두 0이 아닌 계수 $c_1, \ldots, c_N$이 존재하지 않는다.

따라서 벡터 $v_1, \ldots, v_N$는 선형독립이다.

#### 연습 문제 3.2.7의 답

$N$개의 $N$차원 기저벡터 $x_1, x_2, \cdots, x_N$를 행으로 가지는 정방행렬을 $A$라고 하자. 기저벡터는 선형독립이므로 정방행렬 $A$는 풀랭크이고 역행렬이 존재한다.

만약 $x_1, x_2, \cdots, x_N$에 모두 수직인 벡터 $x$가 있다면 다음 식이 성립한다.

$$
\begin{align}
A^Tx = 0
\tag{A.3.55}
\end{align}
$$

$A^T$의 역행렬 $A^{-T}$은 $A$의 역행렬의 전치행렬이므로 존재한다. $A^{-T}$을 위 식의 양변에 곱하면

$$
\begin{align}
A^{-T}A^Tx = x = 0
\tag{A.3.56}
\end{align}
$$ 

따라서 영벡터 이외에 위 조건을 만족하는 벡터 $x$는 존재하지 않는다.

```python
# B 행렬의 SVD
U, S, VT = svd(B, full_matrices=False)

print("[Bv_1 Bv_2]")
print(B @ VT.T)  

print("[s_1u_1 s2_u2]")
print((np.diag(S) @ U.T).T)
```

```python
# C 행렬의 SVD
U, S, VT = svd(C, full_matrices=False)

print("[Cv_1 Cv_2]")
print(C @ VT.T)  

print("[s_1u_1 s2_u2]")
print((np.diag(S) @ U.T).T)
```

#### 연습 문제 3.4.3의 답

```python
from numpy.linalg import svd, eig

A = np.array([[3, -1], [1, 3], [1, 1]])
U, S, VT = svd(A)
U  # 왼쪽 특이벡터
```

```python
w, V = eig(A @ A.T)  # 공분산 행렬
idx = np.argsort(w ** 2)[::-1]  # w^2 크기 순서대로 재배열
w = w[idx]
V = V[:, idx]  
V  # 특이벡터
```
